```python
import os
import subprocess
COMMAND_TEMPLATE = (
    "unzip "
    "{path}"
)
DELETE_TEMPLATE = (
    "rm "
    "-R "
    "{path}"
)
rootdir = os. getcwd()

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if (".DS_Store" not in file) and ("unzip" not in file):
            command = COMMAND_TEMPLATE.format(
            path= os.path.join(subdir, file),
            )
            proc_resp = subprocess.run(
                command.split(" "),
                check=True,
            )
            delete = DELETE_TEMPLATE.format(
            path= subdir,
            )
            proc_resp = subprocess.run(
                delete.split(" "),
                check=True,
            )
```
